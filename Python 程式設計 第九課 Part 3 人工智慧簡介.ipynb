假設我們將所有標籤資料畫在一個平面圖上

---
## 模型複雜度 (Model Complexity)

接下來我們來觀察一下不同的 k 值產生的邊界：

![](https://drive.google.com/uc?export=download&id=1lkiH38fu7kqOaBRT8PagvP7eibJbju_H)

從這張圖我們發現：

1. **N 值越大**，模型產生的邊界就**越平滑 (Smooth)**，這代表我們的模型是一個**低複雜度的模型**
2. **N 值越小**，模型產生的邊界就**越不平滑**，這代表我們的模型是一個**高複雜度的模型**

在一般的情況下：

- 越複雜的模型越容易發生**過適 (overfitting)**，而過適的模型對噪音非常敏感
- 相對的，複雜度越低的模型，會導致**乏適 (underfitting)**，無法捕捉到資料中的規律，導致訓練以及測試資料的準確度會大福下降

此時我們需要做的，是找出一個讓訓練資料與測試資料都能有很高的精準度的 K 值；爲了做到這點，我們可以用不同 K 值以及套用到模型相對應的精準度畫成圖來觀察：

![](https://drive.google.com/uc?export=download&id=1rga0k0yYt6i7iDTlU6BHoVE9yBMNd31T)

````python
import numpy as np
# 產生一個 1 到 8 的串列，以及兩個用來記錄測試與訓練資料集產生的精準度
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
# 將資料集拆解成訓練組與測試組
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# k 值從 1 至 8
for i, k in enumerate(neighbors):
    # 建立一個 k-NN 分類器的物件，放入 knn 變數
    knn = KNeighborsClassifier(n_neighbors=k)
    # 訓練模型
    knn.fit(X_train, y_train)
    # 計算訓練資料集精準度
    train_accuracy[i] = knn.score(X_train, y_train)
    # 計算測試資料集精準度
    test_accuracy[i] = knn.score(X_test, y_test)

# 產生圖表
plt.title('k-NN: Varying Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
````
